## Veri Setinin Betimlenmesi

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
```

```python
df.head()
```

```python
# Veri setinin değişken ve gözlem sayısına erişmek için:
df.shape
```

```python
# Sadece değişken isimlerine erişmek istersek:
df.columns
```

```python
# Betimsel istatistiklere erişme: (Transpozunu almak daha güzel olacaktır.)
df.describe().T
```

* describe() Eksik gözlemleri göz ardı eder ve kategorik değişkenleri dışarıda bırakır. Yani buradaki bütün değişkenler sayısal değişkenlerdir. Burada bütün sayısal değişkenler kategorik değişken olamaz anlamına gelmez. Mesela cinsiyet kategorisi için kadına 1 erkeğe 0 değeri verdiğimizde sayısal değişkenler de kategorik değişken olmuş olurlar. Veri setinin hikayesini bilmek bu yüzden önemlidir.

```python
# hepsini almak için:
df.describe(include = "all").T
# yaparız ancak bu kullanım doğru değildir.
```
