### Çaprazlamalar

```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.boxplot(x = df["total_bill"]);
```

```python
df.describe().T
```

```python
# Hangi günler daha fazla kazanıyoruz?
sns.boxplot(x = "day", y = "total_bill", data = df);
```

```python
df["day"].value_counts()
```

```python
# pazar daha az müşteri olmasına rağmen daha fazla para bırakılmıştır.
```

```python
# sabah mı akşam mı daha çok kazanıyoruz?
sns.boxplot(x = "time", y = "total_bill", data = df );
```

```python
# akşamları daha fazla para kazanılıyor. Yani demek oluyor ki bir restoran işletmecisiysem pazar
# ve akşam yemeklerinde en iyi garsonları çalıştırıp müşterilerin sürekli gelmesini sağlayabilirim.
```

```python
# yemeğe gelen grupların içerisindeki kişi sayısı ile kazanç doğru orantılı mı kontrol edelim? 

sns.boxplot(x = "size", y = "total_bill", data = df);
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = df);
```
