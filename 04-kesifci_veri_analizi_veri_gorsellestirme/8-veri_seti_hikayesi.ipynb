### Dağılım Grafikleri
Elimizdeki sürekli değişkenleri (sayısal veya kategorik değişkenlerin dağılımını) görsel tekniklerle bize sunma imkanı sağlar.


#### Barplot

* Sütun grafikler bize elimizdekikategorik değişkenleri görselleştiirmemize imkan sağlar.

##### Veri seti Hikayesi
- price : dolar cinsinde fiyat (326-18,823)
- carat . ağırlık (0.2-5.01)
- cut : kalite (Fair, Good, Very Good, Premium,Ideal)
- color : renk (from J (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- clarity : temizliği, berraklığı (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF(best))
- x : length in mm (0-10.74)
- y : widthin mm (0-58.9)
- z : depth in mm (0-31.8)
- depth : toplama derinlik yüzdesi = z / mean(x,y) = 2 * z / (x + y) (43-79)
- table : elmasın en geniş noktasına göre genişliği (43-95)

```python
# ordinal tanımlama
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut = df.cut.astype(CategoricalDtype(ordered = True))
```

```python
# Sıralama yanlıştır. Doğrusu şudur:
cut_kategoriler = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
```
