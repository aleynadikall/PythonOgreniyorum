### Kategorik Değişken Özetleri

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

##### Sadece kategorik değişkenler ve özetleri

```python
kat_df = df.select_dtypes(include = ["object"])
# Eğer kategorik değişkenleri önceki kısımlarda gösterildiği gibi kategorik değişkene dönüştürmek istiyorsak bu sefer de tırnak içine category yazmamız gerekecek.
```
