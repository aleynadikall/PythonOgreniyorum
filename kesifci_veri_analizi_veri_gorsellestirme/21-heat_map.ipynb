### Heatmap (Isı Haritası)

#### Veri Seti Hikayesi

```python
import seaborn as sns
flights = sns.load_dataset('flights')
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df['passengers'].describe()
```

```python
# pivot hale getirmeliyiz çünkü daha okunabilir olmasını amaçlıyoruz.

sns.heatmap(df);
# heatmap bizden daha yapısal formda bir dataframe bekler. Bu şekilde hata verir.
```

```python
df = df.pivot("month", "year", "passengers");
```

```python
df
# çok daha okunabilir oldu.
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt = "d");
```

```python
# yukarıdaki heatmapin okunması kafa karışıklığına sebep oluyor. Bu yüzden linewidths kullanalım.
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5);
```

```python
# ölçeklendirmeyi kaldırmak için;
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5, cbar = False);
# bunu yapmak pek önerilmez...
```
