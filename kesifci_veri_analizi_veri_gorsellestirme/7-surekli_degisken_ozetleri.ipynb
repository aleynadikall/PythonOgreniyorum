### Sürekli Değişken Özetleri

```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
df_num = df.select_dtypes(include = ["float64", "int64"])
```

```python
df_num.head()
# sadece sürekli değişkenlere erişmiş olduk
```

```python
df_num.describe().T
#betimsel istatistiklere tüm veri setleri için ulaşmak istersek.
```

```python
df_num["distance"].describe()
#sadece belirli bir değişkenin betimsel istatistik betimlemesine erişmek için:
```

```python
print("Ortalama: " + str(df_num["distance"].mean()))
print("Dolu Gözlem Sayısı: " + str(df_num["distance"].count()))
print("Maksimum Değer: " + str(df_num["distance"].max()))
print("Minimum Değer: " + str(df_num["distance"].min()))
print("Medyan: " + str(df_num["distance"].median()))
print("Standart Sapma: " + str(df_num["distance"].std()))
```

```python
print("Ortalama: " + str(df_num["year"].mean()))
print("Dolu Gözlem Sayısı: " + str(df_num["year"].count()))
print("Maksimum Değer: " + str(df_num["year"].max()))
print("Minimum Değer: " + str(df_num["year"].min()))
print("Medyan: " + str(df_num["year"].median()))
print("Standart Sapma: " + str(df_num["year"].std()))
```

```python
print("Ortalama: " + str(df_num["mass"].mean()))
print("Dolu Gözlem Sayısı: " + str(df_num["mass"].count()))
print("Maksimum Değer: " + str(df_num["mass"].max()))
print("Minimum Değer: " + str(df_num["mass"].min()))
print("Medyan: " + str(df_num["mass"].median()))
print("Standart Sapma: " + str(df_num["mass"].std()))
```
